```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
gut_place_data = pd.read_csv('data/final/gut_place.csv', sep=';')
```

```python
gut_place_data
```

```python
map = folium.Map(location=[55.751244, 37.618423], zoom_start=10, zoom_min=7)
```

```python
for location in gut_place_data.values:
    folium.Marker([location[0], location[1]]).add_to(map)
```

```python
map
```

```python
map.save('gut_place.3k.html')
```
